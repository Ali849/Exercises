## Imort data and libraries

```python
import pandas as pd
import numpy as np
data= pd.read_csv('iris.csv')
```

```python
data
```

### Doing some changing in data 

```python
data.drop('Id',axis=1,inplace=True)
```

```python
data
```

# Set that:
###  0-Iris-setosa
### 1-Iris-versicolor
### 2-Iris-virginica

```python
y=data.values[:,4]

y=np.where(y=='Iris-setosa',0,y)
y=np.where(y=='Iris-versicolor',1,y)
y=np.where(y=='Iris-virginica',2,y)

data['Species']=y
```

```python
data
```

```python
data.to_csv('updated_iris.csv', sep=',', header=True, index=True)
```
